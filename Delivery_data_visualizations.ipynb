```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
df = pd.read_csv('/Delivery Data - datos_2017_2020.csv')
```

```python
df.head(3)
```

```python
df['Ventas'] = df['Ventas'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
df['Costo_envio'] = df['Costo_envio'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
df['Descuento'] = df['Descuento'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
df['Margen'] = df['Margen'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
df['Utilidad'] = df['Utilidad'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
df['Precio_costo'] = df['Precio_costo'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
```

```python

df['fecha_envio'] = pd.to_datetime(df['fecha_envio'], format='%d/%m/%Y')
df_sorted = df.sort_values('fecha_envio')
```

```python
df
```

```python
fig = px.pie(df, values='Cantidad', names='Region',
             title='Grafico de Tarta, porcentaje de Envios y Costos por Region',
             hover_data=['Costo_envio'], labels={'Costo_envio':'Costo_envio'})
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
fig5 = go.Figure()
fig5.add_trace(go.Box(y=df['Descuento'], name='Descuento'))
fig5.add_trace(go.Box(y=df['Utilidad'], name='Utilidad'))

fig5.show()
```

```python
# GrÃ¡fico de Barras Apiladas para Producto y Empaque
fig9 = px.bar(df, x='Producto', y='Ventas', color='Empaque', barmode='stack')
fig9.show()
```

```python
import numpy as np
fig = px.treemap(df, path=['Segmento', 'Region'], values ='Cantidad',
                 color = 'Ventas', color_continuous_scale='RdBu',
                 title='RelaciÃ³n entre Segmento y RegiÃ³n')

# Mostrar el grÃ¡fico
fig.update_layout(
    title_text="GrÃ¡fico de Barras", title_x=0.5,
    annotations=[dict(text="Esta es una descripciÃ³n.", showarrow=False,
                      x=0.5, y=-0.15, xref='paper', yref='paper')]
)
fig.show()
```

```python
fig = px.scatter(df, x='Region', y='Segmento', size='Ventas', color='Segmento',
                 size_max=30, title='RelaciÃ³n entre Segmento y RegiÃ³n')

# ConfiguraciÃ³n adicional para hacer que los marcadores parezcan cÃ­rculos
fig.update_traces(marker=dict(line=dict(color='black', width=1)))
fig.update_layout(showlegend=False)  # Ocultar la leyenda

fig.show()
```
